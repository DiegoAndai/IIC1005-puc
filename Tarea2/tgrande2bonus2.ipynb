```python
import nltk
import gensim
import json
from gensim.models.ldamodel import LdaModel
import pyLDAvis
import pyLDAvis.gensim
```

```python
with open('emol-full.json') as data_file:
    emoldctry=json.load(data_file)
    
with open('la_nacion-full.json') as data_file:
    lndctry=json.load(data_file)

with open('el_mostrador-full.json') as data_file:
    emdctry=json.load(data_file)

with open('la_tercera-full.json') as data_file:
    ltdctry=json.load(data_file)
    
emolLt=[emoldctry,ltdctry]
emLn=[emdctry,lndctry]
```

```python
contentsEmolLt=[]
contentsEmLn=[]
for newspaper in emolLt:
    for article in newspaper:
           contentsEmolLt.append(article['contenido'])
for newspaper in emLn:
    for article in newspaper:
        contentsEmLn.append(article['contenido'])
```

```python
TokensEmolLt=[]
TokensEmLn=[]
stopwords=nltk.corpus.stopwords.words('spanish')
for article in contentsEmolLt:
    tokens=gensim.utils.simple_preprocess(article)
    tokens=[token for token in tokens if token not in stopwords]
    TokensEmolLt.append(tokens)
for article in contentsEmLn:
    tokens=gensim.utils.simple_preprocess(article)
    tokens=[token for token in tokens if token not in stopwords]
    TokensEmLn.append(tokens)
```

```python
labelDictionaryEmolLt=gensim.corpora.dictionary.Dictionary()
labelDictionaryEmLn=gensim.corpora.dictionary.Dictionary()
```

```python
corpusEmolLt=[]
corpusEmLn=[]
for document in TokensEmolLt:
    corpusEmolLt.append(labelDictionaryEmolLt.doc2bow(document,True))
for document in TokensEmLn:
    corpusEmLn.append(labelDictionaryEmLn.doc2bow(document,True))
```

```python
ldaEmolLt=LdaModel(corpusEmolLt, num_topics=5)
```

```python
ldaEmLn=LdaModel(corpusEmLn, num_topics=5)
```

```python
pyLDAvis.enable_notebook()
```

```python
pyLDAvis.gensim.prepare(ldaEmLn, corpusEmLn, labelDictionaryEmLn)
```

```python
pyLDAvis.gensim.prepare(ldaEmolLt, corpusEmolLt, labelDictionaryEmolLt)
```
